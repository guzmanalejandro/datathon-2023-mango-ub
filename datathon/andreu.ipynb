```python
# Make a function to group the outfits described in df2 and join them in df1

def group_outfits(df1, df2):
    '''
    Function to group the outfits described in df2 and join them in df1
    '''
    # Make a copy of df1
    df = df1.copy()
    
    # Make a list of the outfits
    outfits = df2['outfit_id'].unique()
    
    # Make a dictionary with the outfits and the items
    outfit_dict = {}
    for outfit in outfits:
        outfit_dict[outfit] = df2[df2['outfit_id'] == outfit]['cod_articulo'].values
    
    # Make a new column in df1 with the outfits
    df['outfit'] = df['cod_articulo'].map(outfit_dict)
    
    return df
```
