### Applying ARL to learn Color combinations
https://medium.com/mlearning-ai/recommendation-systems-arl-association-rule-learning-bed1a07b5d9a

```python
# Get the COLOR counts for each outfit
# pivot_outfit_colors = pd.pivot_table(merged_data[['cod_outfit', 'cod_color_code', 'des_agrup_color_eng']], index=['cod_outfit'], columns=['des_agrup_color_eng'], aggfunc="count", fill_value = 0.0)
outfit_color_counts = merged_data.groupby(['cod_outfit', 'des_agrup_color_eng'])['des_agrup_color_eng'].count().unstack().fillna(0).map(lambda x: x > 0.0)
outfit_color_counts
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

outfit_frequent_colors = apriori(outfit_color_counts, min_support=0.01, use_colnames=True, max_len = 5).sort_values("support", ascending=False)
outfit_frequent_colors
```

```python
outfit_color_rules = association_rules(outfit_frequent_colors, metric="support", min_threshold=0.01)
outfit_interesting_rules = outfit_color_rules[
	(outfit_color_rules["support"] > 0.1)
    & (outfit_color_rules["confidence"] > 0.1)
    #& (outfit_color_rules["leverage"] > 1)
].sort_values(['confidence', 'support', 'lift'], ascending=False)
outfit_interesting_rules
```

### Combining features to retrieve common norms

```python
pivot_outfit_color_category = pd.pivot_table(merged_data[['cod_outfit', 'cod_modelo_color', 'des_agrup_color_eng', 'des_product_category', 'des_product_family']], index=['cod_outfit'], columns=['des_agrup_color_eng', 'des_product_category', 'des_product_family'], aggfunc="count", fill_value = 0.0).map(lambda x: x > 0.0)
pivot_outfit_color_category = pivot_outfit_color_category.droplevel(0, axis = 1)
outfit_frequent_color_categories = apriori(pivot_outfit_color_category, min_support=0.001, use_colnames=True, max_len = 5).sort_values("support", ascending=False)
outfit_frequent_color_categories
```

```python
outfit_color_categories_rules = association_rules(outfit_frequent_color_categories, metric="support", min_threshold=0.01)
outfit_color_categories_rules = outfit_color_categories_rules.sort_values(['confidence', 'support', 'lift'], ascending=False)
outfit_color_categories_rules
```

### Building an ARL Model

```python
def clothes_in_antecedent(clothes, antecedent):
	'Returns whether the antecedent matches the set of given clothes'
	if len(antecedent) > len(clothes): return False
	for rule in antecedent:
		if clothes[
			(clothes['des_agrup_color_eng'] == rule[0]) & 
			(clothes['des_product_category'] == rule[1])
		].size == 0: return False
	return True

def single_clothe_in_antecedent(clothe, antecedent):
	'Returns whether the antecedent matches the set of given clothes'
	if len(antecedent) > len(clothe): return False
	for rule in antecedent:
		if not ((clothe['des_agrup_color_eng'] == rule[0]) & (clothe['des_product_category'] == rule[1])): return False
	return True
```

```python
def arl_recommender(clothes, rules_df, available_clothes, sort_by = ['confidence', 'support', 'lift']):
	valid_rules = rules_df[rules_df['antecedents'].map(lambda ant: clothes_in_antecedent(clothes, ant))].sort_values(sort_by, ascending=False)
	best_consequence = valid_rules.iloc[0]['consequents']
	possible_clothes = available_clothes[available_clothes.apply(lambda cl: single_clothe_in_antecedent(cl, best_consequence), axis = 1)]
	return possible_clothes
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Function to display product images
def display_product_images(outfit_data):
    # Display images for each product in the outfit
    fig, axes = plt.subplots(1, len(outfit_data), figsize=(15, 5))
    if len(outfit_data) == 1:
        axes = [axes]
    for ax, (index, row) in zip(axes, outfit_data.iterrows()):
        img = mpimg.imread(row['des_filename'])
        ax.imshow(img)
        ax.axis('off')
        ax.set_title(f"{row['des_product_type']}")
    plt.show()
```
